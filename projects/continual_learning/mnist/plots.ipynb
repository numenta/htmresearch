```python
%matplotlib inline
import sys
import numpy as np
from htmresearch.frameworks.pytorch.continual_learning_experiment import \
  BaselineContinualLearningExperiment
import matplotlib.pyplot as plt
from tabulate import tabulate
from IPython.display import HTML
```

```python
sys.argv = ['prog','-c', 'experiments.cfg']
suite = BaselineContinualLearningExperiment()
suite.parse_opt()
suite.parse_cfg()
```

```python
fig, ax = plt.subplots(1, 1)
mean = {}
for exp in suite.get_exps():
    data = suite.get_histories_over_repetitions(exp, 'accuracy', np.mean)
    plt.plot(data, label=exp)
    mean[exp] = np.mean(data)

plt.ylabel('Accuracy')
plt.xlabel('Tasks')
plt.xticks(range(5))
plt.legend()
```

```python
HTML(tabulate(mean.items(), tablefmt='html'))
```
